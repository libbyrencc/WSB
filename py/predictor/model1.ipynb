We now are going to use Long Short Term Memory (LSTM) as our netural network to train our model.

LSTMs are widely used for sequence prediction problems and have proven to be extremely effective. The reason they work so well is because LSTM is able to store past information that is important, and forget the information that is not. LSTM has three gates:

`The input gate`: The input gate adds information to the cell state

`The forget gate`: It removes the information that is no longer required by the model

`The output gate`: Output Gate at LSTM selects the information to be shown as output

reference: https://en.wikipedia.org/wiki/Long_short-term_memory

```python
#transform index
new_data.index = new_data.Date
new_data.drop('Date', axis=1, inplace=True)

 
#creating train and vaild sets
dataset = new_data.values
 
train = dataset[0:511,:]
valid = dataset[511:,:]
 
#converting dataset into X_train and y_train
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

# training model 
X_train, y_train = [], []
for i in range(60,len(train)):
    X_train.append(scaled_data[i-60:i,0])
    y_train.append(scaled_data[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)
 
X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
 
# add LSTM network
model = Sequential([
    layers.LSTM(units = 70, return_sequences = True, input_shape = (X_train.shape[1], 1)), 
    layers.Dropout(0.3),
    layers.LSTM(units = 50),
    layers.Dropout(0.2),
    layers.Dense(units = 1)
    ])

#compile and fit our model
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, y_train, epochs=15, batch_size=32, verbose=2)
 
#using past 60 sets of data from the train data
inputs = new_data[len(new_data) - len(valid) - 60:].values
inputs = inputs.reshape(-1,1)
inputs  = scaler.transform(inputs)

X_test = []
for i in range(60,inputs.shape[0]):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

# making predictions on the test data
closing_price = model.predict(X_test)
closing_price = scaler.inverse_transform(closing_price)
```

```python
# error calculation 
rms = np.sqrt(np.mean(np.power((valid-closing_price),2)))
rms
```
